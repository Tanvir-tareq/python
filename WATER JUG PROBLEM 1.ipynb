```python
'''
Water-Jug solution using Breadth First Search Algorithm
'''

print("Solution for water jug problem")
x_capacity = int(input("Enter Jug 1 capacity: "))
y_capacity = int(input("Enter Jug 2 capacity: "))
end = int(input("Enter target volume: "))

def bfs(start, end, x_capacity, y_capacity):
    from collections import deque

    queue = deque()
    queue.append((start, [start]))
    visited = set()
    visited.add(tuple(start))

    while queue:
        (x, y), path = queue.popleft()

        if x == end or y == end:
            print(f"Found solution in {len(path)-1} steps!")
            return path

        next_states = []
        next_states.append((x_capacity, y))  # Fill jug 1
        next_states.append((x, y_capacity))  # Fill jug 2
        next_states.append((0, y))           # Empty jug 1
        next_states.append((x, 0))           # Empty jug 2

        # Pour from jug 1 to jug 2
        pour = min(x, y_capacity - y)
        next_states.append((x - pour, y + pour))

        # Pour from jug 2 to jug 1
        pour = min(y, x_capacity - x)
        next_states.append((x + pour, y - pour))

        for state in next_states:
            if tuple(state) not in visited:
                visited.add(tuple(state))
                queue.append((state, path + [state]))

    return None

def gcd(a, b):
    if a == 0:
        return b
    return gcd(b % a, a)

# start state: x = 0 , y = 0
start = [0, 0]

if end % gcd(x_capacity, y_capacity) == 0:
    result = bfs(start, end, x_capacity, y_capacity)
    if result:
        print("Optimal solution path:")
        # Create jug labels with capacities
        jug1_label = f"Jug {x_capacity}L"
        jug2_label = f"Jug {y_capacity}L"

        for i, state in enumerate(result):
            print(f"Step {i}: {jug1_label}={state[0]}, {jug2_label}={state[1]}")
    else:
        print("No solution found (but one should exist).")
else:
    print("No solution possible for this combination.")
```
