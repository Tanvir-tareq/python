```python
'''
Water-Jug solution using Breadth First Search Algorithm with Visualization
'''

import os
import time

print("Solution for water jug problem")
x_capacity = int(input("Enter Jug 1 capacity: "))
y_capacity = int(input("Enter Jug 2 capacity: "))
end = int(input("Enter target volume: "))

def bfs_with_visualization(start, end, x_capacity, y_capacity):
    from collections import deque

    queue = deque()
    queue.append((start, [start]))
    visited = set()
    visited.add(tuple(start))

    while queue:
        (x, y), path = queue.popleft()

        if x == end or y == end:
            print(f"\nğŸ¯ Found solution in {len(path)-1} steps!")
            return path

        next_states = []
        next_states.append((x_capacity, y))  # Fill jug 1
        next_states.append((x, y_capacity))  # Fill jug 2
        next_states.append((0, y))           # Empty jug 1
        next_states.append((x, 0))           # Empty jug 2

        # Pour from jug 1 to jug 2
        pour = min(x, y_capacity - y)
        next_states.append((x - pour, y + pour))

        # Pour from jug 2 to jug 1
        pour = min(y, x_capacity - x)
        next_states.append((x + pour, y - pour))

        for state in next_states:
            if tuple(state) not in visited:
                visited.add(tuple(state))
                queue.append((state, path + [state]))

    return None

def visualize_step(x_cap, y_cap, x_amt, y_amt, step_num):
    """Visualize current state of jugs"""
    os.system('cls' if os.name == 'nt' else 'clear')

    print(f"\n{'='*50}")
    print(f"STEP {step_num}: Jug {x_cap}L = {x_amt}/{x_cap} | Jug {y_cap}L = {y_amt}/{y_cap}")
    print(f"{'='*50}")

    max_height = max(x_cap, y_cap)

    # Create visual representation
    jug1_lines = []
    jug2_lines = []

    # Build from top to bottom
    for level in range(max_height, 0, -1):
        # Jug 1
        if level <= x_amt:
            jug1_lines.append(f"  â”‚{'â–ˆ' * 10}â”‚  â† {x_amt}L")
        elif level <= x_cap:
            jug1_lines.append(f"  â”‚{'â–‘' * 10}â”‚")
        else:
            jug1_lines.append(f"  â”‚{' ' * 10}â”‚")

        # Jug 2
        if level <= y_amt:
            jug2_lines.append(f"  â”‚{'â–ˆ' * 10}â”‚  â† {y_amt}L")
        elif level <= y_cap:
            jug2_lines.append(f"  â”‚{'â–‘' * 10}â”‚")
        else:
            jug2_lines.append(f"  â”‚{' ' * 10}â”‚")

    # Print jugs side by side
    for j1, j2 in zip(jug1_lines, jug2_lines):
        print(f"{j1}   {j2}")

    print(f"  â””{'â”€' * 10}â”˜   â””{'â”€' * 10}â”˜")
    print(f"    {x_cap:2d}L          {y_cap:2d}L")
    print(f"{'='*50}")

def gcd(a, b):
    if a == 0:
        return b
    return gcd(b % a, a)

# Main execution
start = [0, 0]

if end % gcd(x_capacity, y_capacity) == 0:
    result = bfs_with_visualization(start, end, x_capacity, y_capacity)

    if result:
        print("\nOptimal solution path:")
        for i, state in enumerate(result):
            visualize_step(x_capacity, y_capacity, state[0], state[1], i)
            if i < len(result) - 1:
                input("\nPress Enter to see next step...")

        print(f"\nâœ… Solution complete! Target volume {end}L reached!")
        print(f"ğŸ“Š Total steps: {len(result)-1}")

        # Show summary
        print(f"\n{'='*50}")
        print("SOLUTION SUMMARY:")
        print(f"{'='*50}")
        for i, (x, y) in enumerate(result):
            action = ""
            if i > 0:
                prev_x, prev_y = result[i-1]
                if x > prev_x and y == prev_y:
                    action = "(Fill Jug 1)"
                elif y > prev_y and x == prev_x:
                    action = "(Fill Jug 2)"
                elif x < prev_x and y == prev_y:
                    action = "(Empty Jug 1)"
                elif y < prev_y and x == prev_x:
                    action = "(Empty Jug 2)"
                elif x < prev_x and y > prev_y:
                    action = f"(Pour {prev_x - x}L from Jug 1 to Jug 2)"
                elif x > prev_x and y < prev_y:
                    action = f"(Pour {prev_y - y}L from Jug 2 to Jug 1)"

            print(f"Step {i:2d}: Jug {x_capacity}L = {x:2d}, Jug {y_capacity}L = {y:2d} {action}")
    else:
        print("No solution found (but one should exist).")
else:
    print("âŒ No solution possible for this combination.")
```
